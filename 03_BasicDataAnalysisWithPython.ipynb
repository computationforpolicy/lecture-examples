# Plotting

Now we'll make some basic plots. Let's look at a histogram of salaries in the city as a whole. First we'll have to get rid of that last row since it has a 'NaN'.

```python
df_salaries['Employee Annual Salary'] = df_salaries['Employee Annual Salary'].str.lstrip('$').astype(float)
```

```python
df_salaries['Employee Annual Salary'].dropna(inplace=True)
```

```python
type(df_salaries['Employee Annual Salary'][0])
```

```python
np.max(df_salaries['Employee Annual Salary'])
```

```python
np.mean(df_salaries['Employee Annual Salary'])
```

## Using matplotlib to make a histogram

```python
n, bins, patches = plt.hist(df_salaries['Employee Annual Salary'], 30)

plt.xlabel('Annual Salary ($)')
plt.ylabel('Number of Employees')
plt.title('Chicago Employee Salaries')
plt.show()
```

## Example of grouping and aggregation operations in pandas

Pandas is useful for its grouping, aggregations, merging and joining capabilities. Here we show an example of that, but note that everything on the homework can be done with loops if you choose. 

```python
grouped = df_salaries.groupby(['Position Title'])['Employee Annual Salary']
aggregated = grouped.agg([np.mean])
```

```python
sorted_jobs = aggregated.sort_values(by='mean',ascending=0)
top_jobs = sorted_jobs.head(20)
```

```python
top_salaries = top_jobs['mean'].values
label_salaries = top_jobs.index.values
```

```python
top_jobs.plot(kind='barh')
```

```python
label_salaries
```

```python
top_salaries
```

## Bar Chart with matplotlib only

```python
positions = np.arange(len(label_salaries)) + 0.4
plt.barh(positions, top_salaries, align='center')
plt.yticks(positions, label_salaries)
plt.xlabel('Annual Salary ($)')
plt.ylabel('Position Title')
plt.show()
```
